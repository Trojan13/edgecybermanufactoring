```python
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
import time
```

```python
time_data = pd.read_csv("data.csv")
time_data= time_data.multiply(1000)
time_data= time_data.round()
time_data
```

```python

from cProfile import label


time_data_descr = time_data.describe()
print(time_data_descr)
fig, ax = plt.subplots(1, 1, figsize=(10, 5))
barlist=ax.bar(time_data_descr.loc["mean"].index, time_data_descr.loc["mean"],
       yerr=time_data_descr.loc["std"], align='center', alpha=0.5, ecolor='black', capsize=10)
barlist[0].set_color('black')
barlist[1].set_color('gray')
barlist[2].set_color('silver')
barlist[3].set_color("gainsboro")
ax.set_ylabel('Mean time (ms)')
ax.legend(loc='best')
plt.show()
plt.close()
```

```python

fig, ax = plt.subplots(1, 1, figsize=(6, 4))
fig1=ax.plot(time_data["Ping Time Edge"].index,time_data["Ping Time Edge"],time_data["Ping Time Cloud"].index,time_data["Ping Time Cloud"])
fig1[0].set_label("Edge")
fig1[1].set_label("Cloud")
ax.set_ylabel('Time (ms)')
ax.set_xlabel('# Testrun')
ax.set_title('Ping Times')
ax.legend(loc='best')
plt.show()
plt.close()
```
